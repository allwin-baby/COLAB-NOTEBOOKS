```python
import keras
from keras import layers
from keras import models
from keras import optimizers
from keras.models import Sequential
```

```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.preprocessing import image
import tensorflow as tf
import matplotlib.pyplot as plt
import numpy as np
import cv2
import os 
```

```python
train = ImageDataGenerator(rescale=1/255)
validation = ImageDataGenerator(rescale=1/255)
```

```python
train_dataset = train.flow_from_directory("/content/drive/MyDrive/dataset/dataset/split/train",     
                                         target_size =(100,100),
                                         batch_size = 32,
                                         class_mode = 'binary')
validation_dataset = validation.flow_from_directory("/content/drive/MyDrive/dataset/dataset/split/val/",     
                                         target_size =(100,100),
                                         batch_size = 32,
                                         class_mode = 'binary')
```

```python
model = Sequential()  
```

```python
model.add(layers.Conv2D(32, (3, 3), activation = 'relu',input_shape=(100, 100,3)))
model.add(layers.MaxPooling2D((2, 2)))
model.add(layers.Conv2D(32, (3, 3), activation = 'relu', ))
model.add(layers.MaxPooling2D((2, 2)))
model.add(layers.Conv2D(64, (3, 3), activation = 'relu', ))
model.add(layers.MaxPooling2D((2, 2)))
model.add(layers.Conv2D(128, (3, 3), activation = 'relu',))
model.add(layers.MaxPooling2D((2, 2)))
model.add(layers.Flatten())
model.add(layers.Dense(units = 512, activation = 'relu'))
model.add(layers.Dense(units = 1, activation = 'sigmoid'))
```

```python
class myCallback(tf.keras.callbacks.Callback):
	def on_epoch_end(self, epoch, logs={}):
		if(logs.get('acc') > .92):
			print("\nReached %2.2f%% accuracy, so stopping training!!" %(.92*100))
			self.model.stop_training = True

# Instantiate a callback object
callbacks = myCallback()
```

```python
model.summary()
model.compile(loss = 'binary_crossentropy',
              optimizer = 'adam',
              metrics = ['acc'])
```

```python
model.fit(train_dataset,
          steps_per_epoch = 108,
          epochs=4,
          validation_data = validation_dataset,
          callbacks=[callbacks])
```

```python
model.save('/content/drive/MyDrive/dataset/swayo_mask.h5') 
```

```python
from keras.models import load_model
model=load_model('/content/drive/MyDrive/dataset/swayo_mask.h5')  
```

```python
import numpy
images = None
total = 0
correct = 0
for i in os.listdir('/content/drive/MyDrive/dataset/dataset/split/test/with_mask'):
    total=total+1
    img = image.load_img("/content/drive/MyDrive/dataset/dataset/split/test/with_mask/"+'//'+ i,target_size =(100,100))
    if (total%100 ==0):
      plt.imshow(img)
      plt.show()
    x= image.img_to_array(img)
    x=  np.expand_dims(x,axis =0)
    images = np.vstack([x])
    val = model.predict(images)
    if(val ==0):
        print('mask')
        correct = correct+1
    if(val ==1):
        print('withoutmask')
print("accuracy  = "+str(correct/total))
```
